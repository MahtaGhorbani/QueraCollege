## اضافه کردن کتابخانه‌های مورد نیاز

```python
import sklearn
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from sklearn.model_selection import train_test_split
```

## خواندن داده‌ها

```python
df = pd.read_csv('wbc.csv')
```

دو ویژگی
`radius_mean` 
و
 `concave points_mean`
را به عنوان داده برای دادن به مدل جدا می‌کنیم.
همچنین
`diagnosis` 
که خوش‌خیم یا بدخیم بودن تومور را مشخص می‌کند را نیز تبدیل به مقادیر عددی می‌کنیم تا برای مدل قابل فهم باشد.

```python
X = df[['radius_mean','concave points_mean']].values
df['diagnosis'].map({'B':0,'M':1})
Y = df['diagnosis']
```

## جدا کردن داده‌ی آموزش و آزمون
تابع `train_test_split` با گرفتن 
`test_size`
به صورت تعداد یا نسبت ورودی و خروجی مدل را به دو قسمت آموزش(`train`) و آزمون(`test`) تقسیم می‌کند.

ولی انتخاب داده‌ها برای قرارگیری در قسمت آموزش و آزمون به صورت تصادفی است که با دادن `random_state` می‌توانیم حالت تصادف آن را تعیین کنیم تا همه افرادی که این کد‌ها را اجرا می‌کنند نتایج یکسانی ببینند.

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify = Y,random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

## آموزش مدل
مدل را برای تعداد همسایگی‌های متفاوت $k$ که در آرایه `K` قرار دارند آموزش می‌دهیم. سپس دقت هرکدام را اندازه‌گیری می‌کنیم و در آرایه‌ی `acc` می‌ریزیم.
- تابع `fit` مدل را برای $X$ و $Y$ داده شده به آن آموزش می‌دهد.
- حال می‌توان با دادن یک دسته ورودی ($X$) با استفاده از تابع `predict` پیش‌بینی مدل برای آنها را بدست آورد.
- تابع `accuracy_score` با گرفتن خروجی درست و خروجی پیش‌بینی شده توسط مدل دقت مدل را برمی‌گرداند.

```python
K = [2,3,4,5,6,7,8,9,10]
acc = []
for k in K:
    knn = KNeighborsClassifier(n_neighbors= k)
    knn.fit(X_train,Y_train)
    y_pred_knn = knn.predict(X_test)
    acc.append(accuracy_score(Y_test,y_pred_knn))
```

## رسم نمودار دقت
حال نمودار دقت در برابر تعداد همسایگی($k$) را رسم می‌کنیم تا ببینیم مدل با چه تعدادی از همسایگی روی این داده بهتر عمل می‌کند.

```python
# Plot k's vs inertias
plt.figure(figsize = (16,6))
plt.plot(K, acc, '-o')
plt.xlabel('number of neighbors, k')
plt.ylabel('accuracy')
plt.xticks(K)
plt.show()
```
